```python
import math
Radius = float(input("Enter how much the radius of the circle is: "))

#Operation to be performed
Circle_area = math.pi * Radius ** 2

#Printing stage
print("The area of a circle whose radius is", Radius, "is:", Circle_area)
```

```python
Assignment #5: Print IP addresses:
```

```python
import ipaddress

#definition stage
net_addr= '192.168.0.3'
net_mask= '255.255.255.0'

#Operation to be performed
Ip_list = ipaddress.IPv4Network(net_addr + '/' + net_mask, strict=False)

#Printing stage
print("All possible ip addresses on said network: ")
for ip_address in Ip_list:
    print(ip_address)
```
